```python
import pandas as pd

# 读取数据
file_path = 'evaluate#.csv'
df = pd.read_csv(file_path)

# # 过滤数据
df = df[df['ARG'] >= 0]
df = df[df['ARG'] <= 100000]
# df = df[(df['num_params'] >= 1)]

# 分组聚合
grouped_df = df.groupby(['pkid', 'layers', 'method'], as_index=False).agg({
    "ARG": 'mean',
    # 'num_params': 'mean',
})

# 定义需要显示的值
values = ["ARG"]

# 创建数据透视表
pivot_df = grouped_df.pivot(index='pkid', columns='method', values=values)

# 设置列的顺序
method_order = ['HeaSolver', 'PenaltySolver', 'ChocoSolver', 'QtoSimplifyDiscardSolver']

# 重新调整列的层次结构
pivot_df = pivot_df.reindex(columns=pd.MultiIndex.from_product([values, method_order]))

# 按 pkid 升序排序
pivot_df = pivot_df.sort_index(axis=0, ascending=True)

```

```python

# 转置 DataFrame
pivot_df = pivot_df.transpose()

pivot_df
```

```python
for row in pivot_df.itertuples(index=False):
    formatted_row = []
    for x in row:
        if isinstance(x, (float, int)):
            if x > 100:
                formatted_row.append(f"{int(x):6d}")  # 大于100保留整数
            elif 10 <= x <= 100:
                formatted_row.append(f"{x:6.1f}")  # 10到100保留1位小数
            else:
                formatted_row.append(f"{x:6.2f}")  # 小于10保留2位小数
        else:
            formatted_row.append(str(x))  # 非数值保持原样
    # 用 & 连接每个值并输出
    print(" & ".join(formatted_row))
```
