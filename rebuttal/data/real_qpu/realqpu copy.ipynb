```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)  # 显示所有行
pd.set_option('display.max_columns', None)  # 显示所有列
```

```python
file_path = 'ibm_kyiv_1#.csv'
df = pd.read_csv(file_path)


metric = 'ARG'
metric2 = 'quantum'
grouped_df0 = df.groupby(['pksid','pkid', 'method'], as_index=False).agg({
    metric: ['mean','std'], 'quantum': ['mean','std'], 'classcial': ['mean','std']
    , 'in_constraints_probs': ['mean','std']
    })

grouped_df0
```

```python
file_path = 'ibm_kyiv_1#.csv'
df = pd.read_csv(file_path)

grouped_df1 = df.groupby(['pksid','pkid', 'method'], as_index=False).agg({
    metric: ['mean','std'], 'quantum': ['mean','std'], 'classcial': ['mean','std']
    , 'in_constraints_probs': ['mean','std']
    })

grouped_df1
```

```python

grouped_df = pd.concat([grouped_df1,grouped_df1])
grouped_df
```

```python
## width 252pt
import numpy as np
import matplotlib.pyplot as plt
## set the figure size and font size
# width of figure in inc
scale = 1
width = 610/64*scale    
# height of figure in inc
height = 100/64*scale
plt.rcParams['figure.figsize'] = (width, height)
plt.rcParams['font.size'] = 10*scale
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'cm'
## set hatch density
plt.rcParams['hatch.linewidth'] = 0.2
plt.rcParams['hatch.color'] = 'black'
plt.rcParams['lines.linewidth'] = 0.7*scale

methods = ['HeaSolver','PenaltySolver','ChocoSolver','QtoSimplifyDiscardSegmentedSolver']
for i,device in enumerate(['ibmq_kyiv','ibm_sherbrooke']):
        axes1 = plt.axes((i*0.3,0,0.3,0.7))
        error_params = dict(elinewidth=0.7*scale,ecolor='black',capsize=1.9*scale,capthick=0.5*scale)
        bar_width = 0.2
        colors = ['#384B70','#6A9C89','#CD5C08','#FFF5E4','#FFF','#FFF']
        hatchs = ['','','','','']
        metric = 'ARG'

        for idx,method in enumerate(methods):
                data = grouped_df[grouped_df.method == method]
                x = np.arange(len(data.pkid))
                y = data[metric]['mean'].to_list()
                plt.bar(x+idx*bar_width-bar_width,y,0.15,color=colors[idx],edgecolor="black",label=method,hatch=hatchs[idx])

        plt.xticks(x+bar_width/2,['F1','K1','S1'])
        plt.xlabel(device)
        plt.yscale('log')
        if i == 0:
                plt.ylabel('ARG')
        else:
                plt.yticks([])
                plt.yticks([],minor=True)
        plt.ylim(0,1e2)

plt.legend(loc='upper left',ncol=4,frameon=False,prop={'size':10*scale},bbox_to_anchor=(0,0,0,0.2),mode="expand",borderaxespad=0)


plt.savefig('realqpuarg.svg',dpi=1200*6.33,bbox_inches='tight')
```

```python
# 绘制散点图，以method 为类别，ARG 为纵坐标，time 为横坐标

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
## set the figure size and font size
# width of figure in inc
scale = 1
width = 252/64*scale    
# height of figure in inc
height = 200/64*scale
plt.rcParams['figure.figsize'] = (width, height)
plt.rcParams['font.size'] = 10*scale
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'cm'
## set hatch density
plt.rcParams['hatch.linewidth'] = 0.2
plt.rcParams['hatch.color'] = 'black'
plt.rcParams['lines.linewidth'] = 0.7*scale
data = pd.read_csv('ibm_kyiv_1#.csv')
data['latency'] = data['classcial']+ data['quantum']
xaxis = 'latency'
yaxis = 'ARG'
markers = ['s','D','^','o','D','*','x','+']
for i,method in enumerate(methods):
    plt.scatter(data[data['method']==method][xaxis]/1000,data[data['method']==method][yaxis],label=method,color=colors[i],marker=markers[i],s=100*scale,alpha=0.7,edgecolor='black',linewidth=0.7*scale)
plt.xlabel(xaxis+' (s)')
plt.ylabel(yaxis)
plt.legend(loc='upper left',ncol=2,frameon=False,prop={'size':10*scale},mode="expand",borderaxespad=0)
plt.savefig('realqpuargtime.svg',dpi=1200*6.33,bbox_inches='tight')
```
