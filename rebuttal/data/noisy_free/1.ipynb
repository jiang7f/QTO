```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# Initialize the figure and axis settings
scale = 1
fig = plt.figure(figsize=(22*scale, 12*scale))
mpl.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42,
    'font.family': 'Times New Roman',
    'font.size': 60,
    'axes.unicode_minus': False,
    'mathtext.fontset': 'custom',
    'mathtext.rm': 'Times New Roman',
    'mathtext.it': 'Times New Roman:italic',
    'mathtext.bf': 'Times New Roman:bold',
    'axes.linewidth': 5,
    'xtick.major.size': 20,
    'xtick.major.width': 5,
    'xtick.minor.size': 10,
    'xtick.minor.width': 3,
    'ytick.major.size': 20,
    'ytick.major.width': 5,
    'ytick.minor.size': 10,
    'ytick.minor.width': 3,
    "lines.markersize" : 35,
    "lines.markeredgewidth" : 4,
    "markers.fillstyle" :  'full',
    'lines.markerfacecolor': '#f8d941',
    'lines.markeredgecolor': 'black',
})
ax = plt.axes((0, 0, 0.65, 0.7))  

# 
df = pd.read_csv("evaluate#.csv")
df = df[df['ARG'] >= 0]
df = df[df['ARG'] <= 100000]
# 
group_cols = ['pkid', 'layers', 'variables', 'constraints', 'method']

#  pbid  10 
df['pbid_group'] = df['pbid'] // 100

#  pbid_group 
averaged_df = df.groupby(group_cols + ['pbid_group']).mean().reset_index()

#  pbid_group
df = averaged_df.drop(columns=['pbid_group'])

# df =df[ df['pkid'] % 4 == 0]

# 
methods = ['HeaSolver', 'PenaltySolver', 'ChocoSolver', 'QtoSimplifyDiscardSolver']
colors = ['#384B70', '#6A9C89', '#B8001F', '#FFF5E4']
markers = ['o', 'o', 'o', 'o']



# 
for idx, (method, color, marker) in enumerate(zip(methods, colors, markers)):
    method_df = df[df['method'] == method]  # 
    ax.scatter(
        method_df["constraints"] / method_df['variables'] + idx * 0.002,  # x
        method_df["ARG"],          # yARG
        label=method,              # 
        c=color,                   # 
        edgecolor='black',linewidth=3,s=1200,
        alpha=0.8                  # 
        ,marker=marker,
    )

# # 
# ax.set_xlabel("Number of Constraints")
# ax.set_ylabel("Approximation Ratio Gap (ARG)")
# ax.set_title("Scatter Plot of Constraints vs ARG by Method")
ax2 = plt.axes((0.75, 0, 0.25, 0.7)) 
#  boxplot 
box_data = [df[df['method'] == method]["ARG"] for method in methods]
boxprops = dict(linewidth=5, color='black')  # 
medianprops = dict(linewidth=5, color='red')  # 

# 
box = ax2.boxplot(
    box_data,
    patch_artist=True,  # 
    boxprops=boxprops,  # 
    medianprops=medianprops,  # 
    showfliers=False,  # 
)

# 
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)  # 
    patch.set_alpha(0.8)  # 
    patch.set_linewidth(5)  # 

#  whiskers  caps 
for whisker, cap in zip(box['whiskers'], box['caps']):
    whisker.set_color('black')
    whisker.set_linewidth(5)  # 
    cap.set_color('black')
    cap.set_linewidth(5)

#  X 
ax2.set_xticks(range(1, len(methods) + 1))
ax2.set_xticklabels(methods, rotation=30, fontsize=18)  #  X 

# 
ax2.legend(
    handles=[plt.Line2D([0], [0], color=color, lw=10) for color in colors],
    labels=methods,
    loc='upper right',
    fontsize=20,
    frameon=False,
)

# 
ax2.set_yscale('log')  #  y 

# 
# ax.legend(title="Methods", fontsize=25, title_fontsize=30, loc='upper right', frameon=True)

# 
ax.grid(True, linestyle='--', linewidth=1.5)
# ax.set_xlim(2.1,14)
# 
ax.tick_params(axis='both', which='major')
# ax.set_ylim(0.001, 100)
# ax.set_ylim(100, 2000)
# ax.set_ylim(100,100000000)
ax.set_yscale('log')
ax.grid(True, linestyle='--', linewidth=1.5, axis='both')
plt.legend(frameon=False, bbox_to_anchor=(0.5, 1.0), loc='lower center', ncol=2)
plt.savefig('d3.svg',bbox_inches='tight')
```
