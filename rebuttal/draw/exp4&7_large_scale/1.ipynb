折线图+拟合

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the data from the CSV file
df = pd.read_csv('more_qubits_without_discard.csv')

# Sort the DataFrame based on the 'variables' column
df_sorted = df.sort_values(by='variables')

# Perform quadratic fitting (degree 2) for num_params vs. variables
coeffs_num_params = np.polyfit(df_sorted['variables'], df_sorted['num_params'], 2)
poly_num_params = np.poly1d(coeffs_num_params)

# Perform quadratic fitting (degree 2) for depth vs. variables
coeffs_depth = np.polyfit(df_sorted['variables'], df_sorted['depth'], 2)
poly_depth = np.poly1d(coeffs_depth)

# Generate the quadratic fit line for num_params
line_fit_num_params = poly_num_params(df_sorted['variables'])

# Generate the quadratic fit line for depth
line_fit_depth = poly_depth(df_sorted['variables'])

# Plot num_params vs. variables with the quadratic fit line
fig, ax1 = plt.subplots(figsize=(22, 12))
ax1.plot(df_sorted['variables'], df_sorted['num_params'], marker='o', linestyle='-', color='#384B70', linewidth=4, markersize=15, markeredgewidth=3, markerfacecolor='#384B70', label='Actual num_params')
ax1.plot(df_sorted['variables'], line_fit_num_params, linestyle='--', color='red', label=f'Fitted Line: $y = {coeffs_num_params[0]:.2f}x^2 + {coeffs_num_params[1]:.2f}x + {coeffs_num_params[2]:.2f}$')
ax1.set_xlabel('variables')
ax1.set_ylabel('num_params')
ax1.grid(True, linestyle='--', linewidth=1.5, axis='y')
ax1.legend()

# Plot depth vs. variables with the quadratic fit line
fig, ax2 = plt.subplots(figsize=(22, 12))
ax2.plot(df_sorted['variables'], df_sorted['depth'], marker='o', linestyle='-', color='#6A9C89', linewidth=4, markersize=15, markeredgewidth=3, markerfacecolor='#6A9C89', label='Actual depth')
ax2.plot(df_sorted['variables'], line_fit_depth, linestyle='--', color='red', label=f'Fitted Line: $y = {coeffs_depth[0]:.2f}x^2 + {coeffs_depth[1]:.2f}x + {coeffs_depth[2]:.2f}$')
ax2.set_xlabel('variables')
ax2.set_ylabel('depth')
ax2.grid(True, linestyle='--', linewidth=1.5, axis='y')
ax2.legend()

# Show the plots
plt.show()

# Print the fitted mathematical expressions
print(f"Fitted equation for num_params: y = {coeffs_num_params[0]:.2f}x^2 + {coeffs_num_params[1]:.2f}x + {coeffs_num_params[2]:.2f}")
print(f"Fitted equation for depth: y = {coeffs_depth[0]:.2f}x^2 + {coeffs_depth[1]:.2f}x + {coeffs_depth[2]:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Read the data from the CSV file
df = pd.read_csv('more_qubits_without_discard.csv')

# Sort the DataFrame based on the 'variables' column
df_sorted = df.sort_values(by='variables')

# Define the exponential function to fit
def exp_func(x, a, b):
    return a * np.exp(b * x)

# Perform exponential fitting for num_params vs. variables
params_num_params, _ = curve_fit(exp_func, df_sorted['variables'], df_sorted['num_params'])

# Perform exponential fitting for depth vs. variables
params_depth, _ = curve_fit(exp_func, df_sorted['variables'], df_sorted['depth'])

# Generate the exponential fit line for num_params
line_fit_num_params = exp_func(df_sorted['variables'], *params_num_params)

# Generate the exponential fit line for depth
line_fit_depth = exp_func(df_sorted['variables'], *params_depth)

# Plot num_params vs. variables with the exponential fit line
fig, ax1 = plt.subplots(figsize=(22, 12))
ax1.plot(df_sorted['variables'], df_sorted['num_params'], marker='o', linestyle='-', color='#384B70', linewidth=4, markersize=15, markeredgewidth=3, markerfacecolor='#384B70', label='Actual num_params')
ax1.plot(df_sorted['variables'], line_fit_num_params, linestyle='--', color='red', label=f'Fitted Line: $y = {params_num_params[0]:.2f} e^{{{params_num_params[1]:.2f}x}}$')
ax1.set_xlabel('variables')
ax1.set_ylabel('num_params')
ax1.grid(True, linestyle='--', linewidth=1.5, axis='y')
ax1.legend()

# Plot depth vs. variables with the exponential fit line
fig, ax2 = plt.subplots(figsize=(22, 12))
ax2.plot(df_sorted['variables'], df_sorted['depth'], marker='o', linestyle='-', color='#6A9C89', linewidth=4, markersize=15, markeredgewidth=3, markerfacecolor='#6A9C89', label='Actual depth')
ax2.plot(df_sorted['variables'], line_fit_depth, linestyle='--', color='red', label=f'Fitted Line: $y = {params_depth[0]:.2f} e^{{{params_depth[1]:.2f}x}}$')
ax2.set_xlabel('variables')
ax2.set_ylabel('depth')
ax2.grid(True, linestyle='--', linewidth=1.5, axis='y')
ax2.legend()

# Show the plots
plt.show()

# Print the fitted mathematical expressions
print(f"Fitted equation for num_params: y = {params_num_params[0]:.2f} e^({params_num_params[1]:.2f}x)")
print(f"Fitted equation for depth: y = {params_depth[0]:.2f} e^({params_depth[1]:.2f}x)")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 读取数据
df = pd.read_csv("more_qubits_without_discard.csv")

# 生成网格数据
m_values = sorted(df["m"].unique())  # 确保 m, n 按升序排列
n_values = sorted(df["n"].unique())

M, N = np.meshgrid(m_values, n_values, indexing="ij")

# 生成 Z 轴数据
Z = np.zeros_like(M, dtype=float)

# 逐行填充 Z 数组
for _, row in df.iterrows():
    m, n, num_params = int(row["m"]), int(row["n"]), row["num_params"]
    Z[m_values.index(m), n_values.index(n)] = num_params  # 确保索引正确

# 创建 3D 图
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")

# 绘制 3D 表面图
ax.plot_surface(M, N, Z, cmap="coolwarm", edgecolor="k", alpha=0.8)

# 轴标签
ax.set_xlabel("m")
ax.set_ylabel("n")
ax.set_zlabel("num_params")
ax.set_title("3D Plot of num_params Growth with m, n")

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 读取数据
df = pd.read_csv("solver_params.csv")

# 生成网格数据
m_values = sorted(df["m"].unique())  # 确保 m, n 按升序排列
n_values = sorted(df["n"].unique())

M, N = np.meshgrid(m_values, n_values, indexing="ij")

# 生成 Z 轴数据
Z = np.zeros_like(M, dtype=float)

# 逐行填充 Z 数组
for _, row in df.iterrows():
    m, n, num_params = int(row["m"]), int(row["n"]), row["num_params"]
    Z[m_values.index(m), n_values.index(n)] = num_params  # 确保索引正确

# 对 Z 数据进行对数转换
Z_log = np.log10(Z)  # 或者使用 np.log(Z)，但要确保 Z > 0

# 创建 3D 图
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")

# 绘制 3D 表面图
ax.plot_surface(M, N, Z_log, cmap="coolwarm", edgecolor="k", alpha=0.8)

# 设置 Z 轴为线性刻度，但已经是对数变换数据
ax.set_zlabel("log(num_params)")
ax.set_title("3D Plot of log(num_params) Growth with m, n")

# 轴标签
ax.set_xlabel("m")
ax.set_ylabel("n")

plt.show()
```
