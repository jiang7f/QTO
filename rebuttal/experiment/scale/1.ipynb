```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 读取数据
df = pd.read_csv("solver_params.csv")

# 生成网格数据
m_values = sorted(df["m"].unique())  # 确保 m, n 按升序排列
n_values = sorted(df["n"].unique())

M, N = np.meshgrid(m_values, n_values, indexing="ij")

# 生成 Z 轴数据
Z = np.zeros_like(M, dtype=float)

# 逐行填充 Z 数组
for _, row in df.iterrows():
    m, n, num_params = int(row["m"]), int(row["n"]), row["num_params"]
    Z[m_values.index(m), n_values.index(n)] = num_params  # 确保索引正确

# 创建 3D 图
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")

# 绘制 3D 表面图
ax.plot_surface(M, N, Z, cmap="coolwarm", edgecolor="k", alpha=0.8)

# 轴标签
ax.set_xlabel("m")
ax.set_ylabel("n")
ax.set_zlabel("num_params")
ax.set_title("3D Plot of num_params Growth with m, n")

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 读取数据
df = pd.read_csv("solver_params.csv")

# 生成网格数据
m_values = sorted(df["m"].unique())  # 确保 m, n 按升序排列
n_values = sorted(df["n"].unique())

M, N = np.meshgrid(m_values, n_values, indexing="ij")

# 生成 Z 轴数据
Z = np.zeros_like(M, dtype=float)

# 逐行填充 Z 数组
for _, row in df.iterrows():
    m, n, num_params = int(row["m"]), int(row["n"]), row["num_params"]
    Z[m_values.index(m), n_values.index(n)] = num_params  # 确保索引正确

# 对 Z 数据进行对数转换
Z_log = np.log10(Z)  # 或者使用 np.log(Z)，但要确保 Z > 0

# 创建 3D 图
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")

# 绘制 3D 表面图
ax.plot_surface(M, N, Z_log, cmap="coolwarm", edgecolor="k", alpha=0.8)

# 设置 Z 轴为线性刻度，但已经是对数变换数据
ax.set_zlabel("log(num_params)")
ax.set_title("3D Plot of log(num_params) Growth with m, n")

# 轴标签
ax.set_xlabel("m")
ax.set_ylabel("n")

plt.show()
```
