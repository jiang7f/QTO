```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 假设数据是从CSV文件加载的，可以通过Pandas加载
methods = ['ChocoSolver', 'HeaSolver', 'PenaltySolver', 'QtoSimplifyDiscardSolver']
colors = ['#384B70', '#6A9C89', '#B8001F', '#FFF5E4']
labels = ['ChocoSolver', 'HeaSolver', 'PenaltySolver', 'QtoSimplifyDiscardSolver']

# 存储所有方法的平均收敛数据
average_convergence = {}

# 读取数据并计算每个方法的平均收敛数据
for method in methods:
    all_data = []
    
    for i in range(10):  # 每种方法有10个文件
        file_name = f"{method}_{i}.csv"
        data = pd.read_csv(file_name)  # 读取CSV数据
        
        # 假设数据的格式是 "Iteration, Cost"
        # ARG = abs((mean_cost - best_cost) / (best_cost + 1e-8))
        best_cost = 12
        cost_values = abs((data['Cost'].values - best_cost) / (best_cost + 1e-8))
        all_data.append(cost_values)
    
    # 计算10个样本的平均收敛值
    average_convergence[method] = np.mean(all_data, axis=0)

# 绘制收敛图
plt.figure(figsize=(12, 8))

# 为每种方法绘制平均收敛图
for i, method in enumerate(methods):
    avg_data = average_convergence[method]
    iterations = np.arange(1, len(avg_data) + 1)
    
    # 每30个点标记一个marker
    mark_every = len(iterations) // 30
    
    plt.plot(iterations, avg_data, label=labels[i], color=colors[i], linewidth=2)
    plt.scatter(iterations[::mark_every], avg_data[::mark_every], color=colors[i], edgecolor='black', marker='o', s=100, zorder=3)

# 设置图例、标题、坐标轴标签
plt.xlabel("Iteration", fontsize=14)
plt.ylabel("ARG", fontsize=14)
plt.title("Convergence of Different Solvers", fontsize=16)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.yscale("log")
# 调整布局并显示
plt.tight_layout()
plt.show()
```
